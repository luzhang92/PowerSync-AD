```python
functions = ['alugo', 'alujs', 'alupy', 'aluphp', 'alurb',
             'aluswift', 'float', 'IOpy', 'markdown2html', 'ocr', 'sentiment']
for function in functions:
    filename = '../tasks/'+function
    with open(filename,'r') as f:
        lines = f.readlines()
    with open(filename+'-0.1f','w') as f:
        for line in lines:
            sepred = line.split()
            init, exec = float(sepred[0]), float(sepred[1])
            init/=1
            exec/=1.1
            f.write(str(round(init)) + '\t' + str(round(exec)) + '\n')
```

```python
# functions = ['alugo', 'alujs', 'alupy', 'aluphp', 'alurb','aluswift', 'float', 'IOpy', 'markdown2html', 'ocr', 'sentiment']
functions = ['sentiment','alugo']

for i in range(len(functions)-1):
    function = functions[i]
    function_d = functions[i+1]
    filename = '../tasks/'+function
    with open(filename,'r') as f:
        lines_f = f.readlines()
        init = int(lines_f[4].split()[0])
        exec = int(lines_f[10].split()[1])
    with open('../tasks/'+function_d,'r') as f:
        lines_d = f.readlines()
        init_base = int(lines_d[4].split()[0])
        exec_base = int(lines_d[10].split()[1])
        init_ratio = init/init_base
        exec_ratio = exec/exec_base

    with open(filename + '-d','w') as f:
        for line in lines_d:
            sepred = line.split()
            init, exec = float(sepred[0]),float(sepred[1])
            init_d,exec_d = round(init_ratio * init), round(exec_ratio * exec)

            f.write(str(init_d)+' '+str(exec_d)+'\n')
```
